```python
# Path to your trained model weights
model_path = '/content/yolov8n.pt'
```

```python
import requests

# Roboflow API key and dataset information
ROBOFLOW_API_KEY = 'aquarium-detection-jiszk'  # Replace with your actual Roboflow API key
workspace_name = 'Aks'       # Replace with your Roboflow workspace name
project_name = 'aquarium-detection-jiszk/1'           # Replace with your Roboflow project name

# API endpoint
upload_url = f"https://api.roboflow.com/{workspace_name}/{project_name}/model-weights?api_key={ROBOFLOW_API_KEY}"

# Open the trained model file in binary mode
with open(model_path, "/content/yolov8n.pt") as model_file:
    response = requests.post(upload_url, files={"file": model_file})

# Check if the upload was successful
if response.status_code == 200:
    print("Model uploaded successfully to Roboflow!")
else:
    print(f"Failed to upload model. Status code: {response.status_code}")
    print(response.json())  # Print any error messages
```
